Introdução a aleatóridade:
Numeros aleatórios são complexos e que são extremamentes dificeis de reproduzir computacionalmente, por isso numeros aleatórios na computação são chamados de pseudo aleatórios, atualmente existem hardwares que transferem a aleatóridade de componentes físicos para o software, atualmente os pseudos aléatorios tem uma aleatóridade tão grande que é impossível/improvável que sua aleatóridade não seja mantido.

```python
import random
print (random.randint(0,100))
```

Agora uma pergunta, caso somamos 2 numeros aleatórios, a aleatóridade aumenta, diminui ou mantém a mesma?


```python
aleat1 = ""
aleat2 = ""
print("Numero aleatório")
for i in range(0,10):
  aleat1+=str(random.randint(0,10))+" "
print(aleat1)
print("Soma de 2 numeros aleatórios")
for i in range(0,10):
  aleat2+=str(random.randint(0,10)+random.randint(0,10))+" "
print(aleat2)
```

É possível ver que apesar de dados muitos pequenos e pouco volume, alguns numeros foram sorteados mais vezes com a soma, o que significa que a aleatóridade diminuiu, será que isso realmente acontece? Vamos utilizar o matplotlib pra simular a aleatóridade.

Biblioteca e instalação do matplotlib no colab.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

Agora com matplotlib instalado, irei criar uma matriz de quantidade de numeros aleatorios e numeros somados.

```python
y = list()
z = list()
for i in range(0,10000):
  y.append(i)
  z.append(0)
for i in range(0,10000):
  ind=0
  for j in range(1,1000):
    ind+=random.randint(0,10)
  z[ind]+=1
plt.plot(y,z)
plt.show()
```
